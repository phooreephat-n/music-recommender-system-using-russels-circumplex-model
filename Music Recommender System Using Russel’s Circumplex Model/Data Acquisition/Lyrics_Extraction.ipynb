## **Get Lyrics from lyricsgenius API**

```python
import lyricsgenius
import time
from requests.exceptions import Timeout
import pandas as pd
```

```python
client_access_token = ''
LyricsGenius = lyricsgenius.Genius(client_access_token)
```

```python
def get_lyrics(song_name, artist_name):
    try:
        song = LyricsGenius.search_song(song_name, artist_name)
        if song is None:
            print(f"Song '{song_name}' by '{artist_name}' not found.")
            return None
        return song.lyrics
    except Timeout:
        print("Timeout, retrying...")
        time.sleep(5)
        return get_lyrics(song_name, artist_name)
    except Exception as e:
        print("Error: ", e)
        return None
```

## Import Dataset

```python
song_df = pd.read_csv('/song_df.csv')
```

## Extract Lyrics

```python
song_df['lyrics'] = song_df.apply(lambda x: get_lyrics(x['Track'], x['Artist']), axis=1)
```
